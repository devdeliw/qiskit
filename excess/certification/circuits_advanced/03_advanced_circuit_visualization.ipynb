```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
```

Alternative Circuit Visualization 

```python
circuit = QuantumCircuit(3, 3)

circuit.x(1)
circuit.h(range(3))
circuit.cx(0, 1)
circuit.measure(range(3), range(3))

print(circuit)
```

```python
circuit.draw()
```

```python
circuit.draw('mpl')
```

```python
circuit.draw('latex')
```

Customizing the Output

Disable Plot Barriers and Reversing Bit Order 

```python
q_a = QuantumRegister(3, name = 'qa')
q_b = QuantumRegister(5, name = 'qb')
c_a = ClassicalRegister(3)
c_b = ClassicalRegister(5)

circuit = QuantumCircuit(q_a, q_b, c_a, c_b)

circuit.x(q_a[1])
circuit.x(q_b[1])
circuit.x(q_b[2])
circuit.x(q_b[4])
circuit.barrier()
circuit.h(q_a)
circuit.barrier(q_a)
circuit.h(q_b)
circuit.cswap(q_b[0], q_b[1], q_b[2])
circuit.cswap(q_b[2], q_b[3], q_b[4])
circuit.cswap(q_b[3], q_b[4], q_b[0])
circuit.barrier(q_b)
circuit.measure(q_a, c_a)
circuit.measure(q_b, c_b)

circuit.draw('mpl')
```

```python
circuit.draw('mpl', reverse_bits=True)
```

```python
circuit.draw('mpl', plot_barriers=False)
```

```python
circuit.draw('mpl', reverse_bits=True, plot_barriers=False)
```

```python
style = {'backgroundcolor': 'lightgreen'}

circuit.draw('mpl', style = style)
```

```python
circuit.draw('mpl', style = style, scale = 0.5)
```

```python
from qiskit.visualization import circuit_drawer

circuit_drawer(circuit, output = 'mpl')
```
