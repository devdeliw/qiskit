```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
```

```python
bell = QuantumCircuit(2, 2)

bell.h(0)
bell.cx(0, 1)
bell.barrier(range(2))
bell.measure(range(2), range(2))

backend = AerSimulator()
bell_compiled = transpile(bell, backend)
job_sim1 = backend.run(bell_compiled, shots = 1000)
job_sim2 = backend.run(bell_compiled, shots = 1000)
job_result1 = job_sim1.result().get_counts()
job_result2 = job_sim2.result().get_counts()

legend = ['sim1', 'sim2']

plot_histogram([job_result1, job_result2], legend = legend)
```
